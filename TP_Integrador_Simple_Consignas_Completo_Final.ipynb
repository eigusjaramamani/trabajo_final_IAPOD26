## Datos del trabajo

Vamos a usar una lista de diccionarios.  
Cada diccionario representa a un estudiante.

```python
estudiantes = [
    {"nombre": "Ana", "edad": 22, "trabaja": "Sí", "horas_estudio": 5, "nota": 8, "usa_ia": "Sí"},
    {"nombre": "Juan", "edad": 28, "trabaja": "Sí", "horas_estudio": 2, "nota": 5, "usa_ia": "No"},
    {"nombre": "Sofía", "edad": 20, "trabaja": "No", "horas_estudio": 6, "nota": 9, "usa_ia": "Sí"},
    {"nombre": "Pedro", "edad": 24, "trabaja": "No", "horas_estudio": 3, "nota": 6, "usa_ia": "No"},
    {"nombre": "Lucía", "edad": 31, "trabaja": "Sí", "horas_estudio": 4, "nota": 7, "usa_ia": "Sí"},
    {"nombre": "Martín", "edad": 19, "trabaja": "No", "horas_estudio": 2, "nota": 4, "usa_ia": "No"},
    {"nombre": "Carla", "edad": 26, "trabaja": "Sí", "horas_estudio": 7, "nota": 9, "usa_ia": "Sí"},
    {"nombre": "Diego", "edad": 23, "trabaja": "No", "horas_estudio": 4, "nota": 7, "usa_ia": "Sí"},
]
```

## Actividad 1: convertir los datos en una tabla

1.   Elemento de la lista
2.   Elemento de la lista



Importá pandas y convertí la lista `estudiantes` en un DataFrame llamado `df`.

Luego mostrá la tabla.

```python
# Escribí tu código acá
import pandas as pd
df = pd.DataFrame(estudiantes)

#Para mostrar el dataframe
print(df)
```

## Actividad 2: calcular indicadores generales

Calculá y mostrá:

1. cantidad total de estudiantes;
2. promedio de edad;
3. promedio de nota;
4. promedio de horas de estudio.

```python
# Escribí tu código acá
#Inciso 1 Cantidad de estudiantes
cantidad_estudiantes = len(df)
print("Cantidad total de estudiantes:", cantidad_estudiantes)


#Inciso 2 Promedio edad
promedio_edad = df['edad'].mean()
print("Promedio de edad:", promedio_edad)

#Inciso 3 Promedio nota
promedio_nota = df['nota'].mean()
print("Promedio de nota:", promedio_nota)

#Inciso 4 promedio de horas de estudio
promedio_horas_estudio = df['horas_estudio'].mean()
print("Promedio de horas de estudio:", promedio_horas_estudio)
```

## Actividad 3: aprobados y desaprobados

Consideramos aprobado a quien tiene nota igual o mayor a 6.

Creá una nueva columna llamada `estado`, con los valores:

- `Aprobado`
- `Desaprobado`

Luego contá cuántos estudiantes aprobaron y cuántos desaprobaron.

```python
# Escribí tu código acá
#Creando la columna estado
df['estado'] = df['nota'].apply(lambda x: 'Aprobado' if x >= 6 else 'Desaprobado')

#Contando la cantidad de estudiantes aprobados y desaprobados
aprobados = df[df['estado'] == 'Aprobado'].shape[0]
desaprobados = df[df['estado'] == 'Desaprobado'].shape[0]

print("Cantidad de estudiantes aprobados:", aprobados)
print("Cantidad de estudiantes desaprobados:", desaprobados)
```

## Actividad 4: trabajo y rendimiento

Separá los estudiantes en dos grupos:

- quienes trabajan;
- quienes no trabajan.

Luego calculá el promedio de nota de cada grupo.

```python
# Escribí tu código acá
# Separando los estudiantes en dos grupos
trabajan = df[df["trabaja"] == "Sí"]
no_trabajan = df[df["trabaja"] == "No"]

# Promedio de nota de cada grupo
promedio_trabajan = trabajan["nota"].mean()
promedio_no_trabajan = no_trabajan["nota"].mean()

print("Promedio de nota de quienes trabajan:", promedio_trabajan)
print("Promedio de nota de quienes no trabajan:", promedio_no_trabajan)
```

## Actividad 5: uso de IA

Contá:

- cuántos estudiantes usan IA;
- cuántos estudiantes no usan IA.

```python
# Escribí tu código acá
# Contar estudiantes que usan IA y los que no
usan_ia = (df["usa_ia"] == "Sí").sum()
no_usan_ia = (df["usa_ia"] == "No").sum()

print("Usan IA:", usan_ia)
print("No usan IA:", no_usan_ia)
```

## Actividad 6: nota más alta y nota más baja

Mostrá:

- el nombre del estudiante con la nota más alta;
- el nombre del estudiante con la nota más baja.

```python
# Escribí tu código acá
# Estudiante con la nota más alta
mayor = df.loc[df["nota"].idxmax()]

# Estudiante con la nota más baja
menor = df.loc[df["nota"].idxmin()]

print("Nota más alta:")
print(mayor["nombre"], "-", mayor["nota"])

print("Nota más baja:")
print(menor["nombre"], "-", menor["nota"])
```

## Actividad 7: gráficos

Realizá al menos dos gráficos:

1. gráfico de barras con las notas de cada estudiante;
2. gráfico de dispersión con horas de estudio y nota final.

Podés usar `matplotlib`.

```python
# Escribí tu código acá
#Importar libreria
import matplotlib.pyplot as plt
```

```python
#Gráfico de barras con las notas de cada estudiante
plt.figure(figsize=(8,5))

plt.bar(df["nombre"], df["nota"])

plt.title("Notas de los estudiantes")
plt.xlabel("Estudiantes")
plt.ylabel("Nota")

plt.show()
```

```python
#Gráfico de dispersión
plt.figure(figsize=(8,5))

plt.scatter(df["horas_estudio"], df["nota"])

plt.title("Horas de estudio y nota")
plt.xlabel("Horas de estudio")
plt.ylabel("Nota")

plt.show()
```
